```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree 
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# model=DecisionTreeClassifier(max_depth=3) post pruning 
model=DecisionTreeClassifier() #unprunned model 
model.fit(x_train,y_train)
predict_train=model.predict(x_train)
predict_test=model.predict(x_test)
```

```python
print("accuracy score of train data:->",accuracy_score(predict_train,y_train))
print("accuracy score of test data:->",accuracy_score(predict_test,y_test))
```

```python
# lets work on ensemble method 
#there are three ensemble method that is bagging ,boosting and staking 
```

```python
#bagging 
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
base_model=DecisionTreeClassifier()
Bagging_model=BaggingClassifier(base_model,n_estimators=50,random_state=42)
Bagging_model.fit(x_train,y_train)
predict_train=Bagging_model.predict(x_train)
predict_test=Bagging_model.predict(x_test)

print("accuracy score of train data:->",accuracy_score(predict_train,y_train))
print("accuracy score of test data:->",accuracy_score(predict_test,y_test))
```

```python
#random forest
random_forest_model=RandomForestClassifier()
random_forest_model.fit(x_train,y_train)
predict_train=random_forest_model.predict(x_train)
predict_test=random_forest_model.predict(x_test)
print("accuracy score of train data:->",accuracy_score(predict_train,y_train))
print("accuracy score of test data:->",accuracy_score(predict_test,y_test))
```

```python
#boosting method
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
model1=model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model1.fit(x_train,y_train)
predict_train=model1.predict(x_train)
predict_test=model1.predict(x_test)
print("accuracy score of train data:->",accuracy_score(predict_train,y_train))
print("accuracy score of test data:->",accuracy_score(predict_test,y_test))
```

```python
#adaboosting method
model2=AdaBoostClassifier(base_model,n_estimators=50,random_state=42)
model2.fit(x_train,y_train)
predict_train=model2.predict(x_train)
predict_test=model2.predict(x_test)
print("accuracy score of train data:->",accuracy_score(predict_train,y_train))
print("accuracy score of test data:->",accuracy_score(predict_test,y_test))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
data=pd.read_csv(r"C:\Users\DELL\Downloads\household_power_consumption.txt",sep=';',low_memory=True)
#data["Global_active_power"]=pd.to_numeric(data["Global_active_power"])

date=data['Date']

data['real_formate_date']=pd.to_datetime(date,dayfirst=False)
data['year']=data["real_formate_date"].dt.year
x=np.array(data["year"]).reshape(-1,1)
y=np.array(data["Global_active_power"])
y=y[:6000]
x=x[:6000]

data.isnull().sum()
data.drop('Sub_metering_3',axis=1,inplace=True)
data.isnull().sum()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=30)
model=LinearRegression()
model.fit(x_train,y_train)
prid=model.predict(x_test)
plt.scatter(x_train,y_train,color='blue',label="test data",linewidth=1)
plt.plot(x_test,prid,color='red',label='regression line',linewidth=2)
#plt.plot(x_test,y_test,color='green')
plt.legend()
plt.show()
# print(mean_absolute_error(y_test,prid))
# print(mean_squared_error(y_test,prid))
# print(r2_score(y_test,prid))






```

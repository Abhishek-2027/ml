```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

data=pd.read_csv(r"C:\Users\DELL\Downloads\score.csv")
```

```python
data
x=np.array(data['Hours'])
y=np.array(data['Scores'])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
model=LinearRegression()
model.fit(x_train.reshape(-1,1),y_train)
predicted=model.predict(x_test.reshape(-1,1))
print("intercept=",model.intercept_)
print("slope=",model.coef_[0])
print(predicted)
plt.scatter(x_train,y_train,color="red",label="train score")

plt.plot(x_train,model.predict(x_train.reshape(-1,1)),color="blue",label="regression line",marker="v")
plt.plot(x_test,predicted,color="yellow",label="predicted line",linestyle="dotted",linewidth=2)
plt.scatter(x_test,y_test,color="green",label="test data")
plt.legend()
plt.show()
mae=mean_absolute_error(y_test,predicted)
print("mae=",mae)
mse=mean_squared_error(y_test,predicted)
print('mse=',mse)
print('rmse=',np.sqrt(mse))
print('r^2=',r2_score(y_test,predicted))

# mae = mean_absolute_error(y_test, predicted)
# mse = mean_squared_error(y_test,predicted)
# rmse = np.sqrt(mse)
# r2 = r2_score(y_test,predicted)

         
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import  matplotlib.pyplot as plt
data=pd.read_csv(r"C:\Users\DELL\Downloads\amazon_sales_dataset.csv")

x=np.array(data['Year']).reshape(-1,1)
y=np.array(data['Total_MarkDown']).reshape(-1,1)
data.drop(['Type','Temperature'],inplace=True,axis=1)
# data.rename({"Unnamed:3":"a"},axis="columns",inplace=True)
data=data.loc[ : ,~data.columns.str.contains('Unnamed')]
data=data.rename(columns={"Size":"size"})
data
print(data)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=30)
model=LinearRegression()
model.fit(x_train,y_train)
prid=model.predict(x_test)
plt.scatter(x_train,y_train,color='blue',label="test data",linewidth=1)
plt.plot(x_test,prid,color='red',label='regression line',linewidth=2)
#plt.plot(x_test,y_test,color='green')
plt.show()
print(mean_absolute_error(y_test,prid))
print(mean_squared_error(y_test,prid))
print(r2_score(y_test,prid))

```

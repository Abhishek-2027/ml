```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
x=np.array([100,150,200,250,300]).reshape(-1,1) #size of house in square feet
y=np.array([25000,35000,45000,55000,65000]) #price of house 
model=LinearRegression()
model.fit(x,y)
new_x=np.array([115,222]).reshape(-1,1)
new_y=model.predict(new_x)# npredict price of  new house size
print(new_y)
print("intercept = ",model.intercept_)
print("slope =",model.coef_)
plt.scatter(x,y,color='green',label="data")
plt.plot(x,model.predict(x),label='regression line',color='red')
plt.scatter(new_x,new_y,color='blue',label='predict value')
plt.xlabel('size of house in square fit')
plt.ylabel("price of house ")
#plt.legend(loc='lower center')
plt.legend()
plt.show()
```
